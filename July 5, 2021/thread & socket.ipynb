# thread & socket

os-스케줄러. 어느 순간에 어떤 작업(task)을 실행할지 결정  
  
유니 태스킹: 한 순간에 한 task만 실행  
멀티 태스킹: 한 순간에 여러 task 실행. 하나의 프로그램을 task 단위로 쪼개서 구현해야 함.  
  
task 구현 => process, thread

<쓰레드 생성>  
  
th = threading.Thread(target=param1, args=param2)  
  
param1:이 쓰레드가 실행할 코드를 함수로 만들어서 파라메터로 지정  
  
param2:param1에서 지정한 함수가 파라메터가 있다면 param2에서 그 인자를 튜플형태로 전달  
  
<쓰레드 시작>  
th.start(): 쓰레드를 ready 상태로 만듬  
  
<쓰레드 대기상태 만들기>  
time.sleep(t): 실행중인 쓰레드를 t시간만큼 대기(wait)상태로 만듬.   
t시간이 끝나면 다시 ready상태로 가서 줄선다.  
자기 순서 다시 돌아오면 cpu획득해서 다시 run 상태 됨

```python
import threading, time

def f():
    for i in range(0, 26):
        print('th:', i)
        time.sleep(1)

def main():
    th = threading.Thread(target=f)#쓰레드 생성
    th.start()
    s = 'abcdefghijklmnopqrstuvwxyz'
    for i in s:
        print('main:', i)
        time.sleep(0.3)
```

```python
main()
```

동기화 처리  
  
공유자원:여러개의 쓰레드가 공유해서 사용하는 데이터나 자원. 동기화처리 필요  
동기화: 배타적 접근 제공. 공유자원에 동시에 접근하는 쓰레드들을 관리한다(줄세우기)  

```python
import threading, time

#전역변수: 함수 밖에 정의. 모든 함수가 공유
flag = False  

def my_input(flag):  # 작업1
    while True:
        msg = input('msg:')#종료 메시지:/end
        if msg=='/end':
            print('채팅을 종료함')
            flag = True
            break
        print('my input:', msg)
    print('th1 종료')


def my_output():  # 작업2
    s = ['aaa', 'bbb', 'ccc', 'ddd', 'eee', 'fff', 'ggg', 'hhh', 'iii', 'jjj']
    cnt=0
    while True:
        if flag:
            break
        print('server msg:', s[cnt%10])
        cnt+=1
        time.sleep(1)
    print('th2 종료')


def main():
    th1 = threading.Thread(target=my_input)
    th1.start()

    th2 = threading.Thread(target=my_output)
    th2.start()

    print('main 종료')

main()
```

네트워크(소켓통신)  
  
주소   
맥주소: 하드웨어 주소. 랜카드 고유번호. 로컬 네트워크 통신.  
논리주소: 인터넷에서 사용하는 주소.   
  
IPv4: 32비트 주소  
  
A class : netID(1B):hostID(3B)  => 동네이름8비트로 표현(총 256개의동네)/컴퓨터한대는 24비트 표현(한 네트워크를 구성하는 컴퓨터의 개수는 2의 24승개)  
동네 개수는 작지만 한 네트워크가 갖는 컴퓨터의 개수가 크다 => 큰 네트워크 구성에 적합. 대규모 네트워크 구성에 적합  
  
B class :  netID(2B):hostID(2B)  => 네트워크 2의 16승개 / 한 네트워크에 포함된 컴퓨터는 2의 16승개 => 중형 네트워크 구축에 적합  
  
C class :  netID(3B):hostID(1B)  => 네트워크 2의 24승개 / 한 네트워크에 포함된 컴퓨터는 2의 8승개(256개). 실생활에서 많이 접하는 주소  
  
서브넷마스크  
1) netID를 확인하기 위한 마스크=>ex>A클래스:255.0.0.0 / B클래스: 255.255.0.0 / C클래스: 255.255.255.0  
2) 동네를 더 쪼개서 사용하고 자 할때 사용.   
C클래스에서 netID는 3바이트로 표현. 1바이트는 host 주소를 표현하는에 이 비트를 더 쪼개서 작은 동네를 표현할 수 있음
예로 hostID 8비트중 2비트를 서브넷으로 지정할 경우 4개의 하위 네트워크를 구축할 수있음. 이 subnet에는 컴퓨터 2의 6승개(64)로 구성됨  
=>255.255.255.192  
  
프로토콜: 통신규약(약속)  
  
네트워크 계층>  
1) 응용계층: 데이터를 받아서 어플리케이션에서 사용. 포트번호(네트워크 앱을 구분하는 고유한 번호)   
 
2) 전송계층(TCP/UDP): 데이터를 목적지에 전달하기 위해 길을 수립(라우팅). 길을 수립. 패킷을 전달.  패킷의 수신, 오류 체크  
 
3) ip 계층(IP/ARP/RARP): 전송할 데이터를 패킷 단위로 분리(단편화). 패킷에 목적지, 보낸는 주소 논리주소로 설정. 주소 변환(논리주소를 맥주소 또는 맥주소를 논리주소로 변환)  
  
4) 물리계층: 랜선으로 직접 물려있는 통신.  맥주소로 로컬 통신.  
  
IP: 컴퓨터의 주소  
PORT: IP가 가르키는 컴퓨터 내에서 네트워크 프로그램을 구분할 수 있는 유일한 번호  
  
1.TCP  : 신뢰성 있고 연결지향적이다  
    한번 수립한 루트를 연결 끊을때까지 계속 사용함  
    상대방이 패킷을 받았는지 확인, 순서, 오류 체크  
    속도는 느리다  
  
2.UDP  : 신뢰성 없고 비연결지향적이다  
    패킷 전달할때마다 새 길을 수립하고 순서, 오류 등 체크하지 않음  
    속도가 빠름. 전사적으로 메시지 전달하는 작업이나 브로드캐스팅에 적합함  

```python
#서버 프로그램
import socket, time

host = '192.168.103.55' #서버 주소
port = 3333             #서버 포트번호

#서버소켓오픈(대문을 열어둠): socket.AF_INET:주소종류(ip) 
#socket.SOCK_STREAM:통신종류.(UDP, TCP). socket.SOCK_STREAM는 TCP통신
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

#여러번 ip,port를 바인드하면 에러나므로 그걸 방지하기 위한 설정
server_socket.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)

#server socket에 ip와 port 붙여줌(바인드)
server_socket.bind((host, port))

#클라이언트 접속 준비 완료
server_socket.listen()

print('echo server start')

#accept(): 클라이언트 접속 기다리며 대기. 
#클라이언트가 접속하면 서버와 클라이언트가 1:1통신할 소켓(빨대)와, 
#접속한 클라이언트의 주소를 반환
client_soc, addr = server_socket.accept()

print('connected client addr:', addr)

#recv(메시지크기): 소켓에서 크기만큼 읽음. 
#소켓에 읽을 데이터 없으면 생길때까지 기다림
data = client_soc.recv(100) 
msg = data.decode()#읽은 데이터 디코딩
print('recv msg:', msg)
client_soc.sendall(msg.encode(encoding='utf-8'))#클라이언트에 에코 메시지 전송

time.sleep(5)
server_socket.close()
```

위 코드를 한명의 클라이언트가 메시지 여러번 주고 받을 수 있게 수정하시오.  
클라이언트가 '/end' 입력하면 서버, 클라이언트 모두 종료  

```python
#서버 프로그램
import socket, time

host = '192.168.103.55' #서버 주소
port = 3333             #서버 포트번호

#서버소켓오픈(대문을 열어둠): socket.AF_INET:주소종류(ip) 
#socket.SOCK_STREAM:통신종류.(UDP, TCP). socket.SOCK_STREAM는 TCP통신
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

#여러번 ip,port를 바인드하면 에러나므로 그걸 방지하기 위한 설정
server_socket.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)

#server socket에 ip와 port 붙여줌(바인드)
server_socket.bind((host, port))

#클라이언트 접속 준비 완료
server_socket.listen()

print('echo server start')

client_soc, addr = server_socket.accept()

print('connected client addr:', addr)

while True:
    data = client_soc.recv(100) 
    msg = data.decode()#읽은 데이터 디코딩
    print('recv msg:', msg)
    client_soc.sendall(msg.encode(encoding='utf-8'))#클라이언트에 에코 메시지 전송
    if msg=='/end':
        break

time.sleep(5)
print('서버 종료')
server_socket.close()
```

여러 클라이언트가 에코 서버에 접속할 수 있도록 서버 프로그램을 수정하시오

```python
#서버 프로그램
import socket, time, threading

def f1(soc):
    while True:
        data = soc.recv(100) 
        msg = data.decode()#읽은 데이터 디코딩
        print('recv msg:', msg)
        soc.sendall(msg.encode(encoding='utf-8'))#클라이언트에 에코 메시지 전송
        if msg=='/end':
            break

def main():
    host = '192.168.103.55' #서버 주소
    port = 3333             #서버 포트번호

    server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server_socket.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
    server_socket.bind((host, port))
    server_socket.listen()
    print('echo server start')
    
    while True:
        client_soc, addr = server_socket.accept()#접속대기
        th = threading.Thread(target=f1, args=(client_soc,))
        th.start()
        print('connected client addr:', addr)
    

    time.sleep(5)
    print('서버 종료')
    server_socket.close()
```

```python
main()
```

채팅 프로그램

```python
#채팅서버
import socket, threading

members=[]#접속한 클라이언트 소켓. 채팅방

def sendMsgAll(msg):#인자로 받은 메시지를 채팅방에 있는 모든 사람에 전송
    print('sendMsgAll:', msg)
    for i in members:
        i.sendall(msg.encode(encoding='utf-8'))
            
#쓰레드 작업. 클라이언트 하나당 하나 생성됨. 소켓으로 연결된 클라이언트가 전송한 메시지를 받아서 전체에 전달
def f1(p, soc):
    id = 'member'+str(p)
    msg = id+'님이 입장하셨습니다'
    sendMsgAll(msg)
    
    while True:
        msg = soc.recv(100)#자기가 담당한 클라이언트가 메시지 보내면 받는다
        msg = msg.decode()
        if msg=='/end':           
            soc.sendall(msg.encode(encoding='utf-8'))#담당하는 클라이언트에 /end보냄
            members.remove(soc)#채팅방에서 제거
            msg = id+'님이 퇴장하셨습니다'
            sendMsgAll(msg)
            break
        msg = id+': '+msg
        sendMsgAll(msg)
            
def main():
    host = '192.168.103.55' #서버 주소
    port = 3334             #서버 포트번호

    server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server_socket.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
    server_socket.bind((host, port))
    server_socket.listen()
    print('chatting server start')
    
    while True:
        client_soc, addr = server_socket.accept()#클라이언트 기다림
        members.append(client_soc)
        print('connected client addr:', addr)
        #쓰레드 생성
        th = threading.Thread(target=f1, args=(addr[1], client_soc))
        th.start()
        
    server_socket.close()
    
main()
```

```python
#채팅 클라이언트
#쓰레드로 작업을 분리
#th1: 서버가 전송한 메시지 받아서 출력
#th2: 메시지를 키보드 입력받아서 서버로 전송. /end 입력하면 종료

import socket, threading

soc = None#전역변수. 서버와 통신할 소켓. sendMsg(), recvMsg() 모두 사용


def sendMsg():#th1
    while True:
        msg = input('msg:')
        soc.sendall(msg.encode(encoding='utf-8'))
        if msg == '/end':
            break


def recvMsg():#th2
    while True:
        data = soc.recv(100)
        msg = data.decode()  # 읽은 데이터 디코딩
        print(msg)
        if msg == '/end':
            break

    soc.close()


def main():
    global soc
    server_ip = '192.168.103.55'
    server_port = 3334

    soc = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    soc.connect((server_ip, server_port))

    th1 = threading.Thread(target=sendMsg)
    th1.start()

    th2 = threading.Thread(target=recvMsg)
    th2.start()

main()
```
